```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Remilyn29/Acti-1/main/SALARY_DATA.csv')
```

```python
print(df)
```

```python
salary = df['Salary'].to_numpy()
```

```python
mean_salary = np.mean(salary)
```

```python
print(mean_salary)
```

```python
median_salary = np.median(salary)
```

```python
print(median_salary)
```

```python
std_salary = np.std(salary)
```

```python
print(std_salary)
```

```python
total_salary = np.sum(salary)
```

```python
print(total_salary)
```

```python
sns.histplot(df['Salary'])
```

```python
sns.catplot(data=df,col = "Name",x = "Salary",kind = "count")
```

```python
df_h = df.groupby(['Name' , 'Salary']).size().head(10)
```

```python
(df_h.head())
```

```python
df_h.plot(kind='bar')
plt.title('total salary')
plt.xlabel('Name')
plt.ylabel('Highest salary')
```

```python
top_5 = df_h.sort_values(ascending = True).head(5)
```

```python
top_5.plot(kind='bar')
plt.title('total salary')
plt.xlabel('Name')
plt.ylabel('Highest salary')
```
